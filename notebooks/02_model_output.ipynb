# Model Output Demo
This notebook demonstrates loading the trained artifacts and running predictions on sample rows from the dataset.

```python
import joblib
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Paths (notebook located in notebooks/ so go up one level)
MODEL_DIR = '../models'
DATA_PATH = '../data/vendor_proposals.csv'

model = joblib.load(MODEL_DIR + '/vendor_model.pkl')
scaler = joblib.load(MODEL_DIR + '/scaler.pkl')
features = joblib.load(MODEL_DIR + '/features.pkl')

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
# Prepare a small sample for prediction
sample = df.head(5).copy()
y_true = sample['vendor_score'].values
X = sample.drop(columns=['vendor_id', 'vendor_score'])
X = pd.get_dummies(X)
# Align columns to training features
X = X.reindex(columns=features, fill_value=0)
# Scale numeric columns (scaler expects same ordering that was used during training)
X_scaled = scaler.transform(X)
preds = model.predict(X_scaled)
result = sample[['vendor_id']].copy()
result['vendor_score_true'] = y_true
result['vendor_score_pred'] = np.round(preds, 3)
result
```
